## Goal Detection

```python
count = 0
goals_count = 0
goal4A = 0
goal4B = 0

def Goal_Detection(result, frame_threshold = 350):
  global count
  global goals_count
  global goal4A
  global goal4B
  #print(count)
  if count > 0:
    count += 1
    if count >= frame_threshold:
      count = 0
    return "", goals_count, goal4A, goal4B

  min_dis = 100000000000000000000000000000000000000000000
  ball=[]
  A = []
  B = []
  team = []
  distances = []
  ball_xc , ball_yc= 0, 0
  mat = result.data.cpu()
  t = np.array(mat)
  num_players = 0
  message = ''
  goal = []

  if len(t) == 0:
    return "", goals_count, goal4A, goal4B

  i = 6
  for box in t:
    if len(box) == 6:
      i = 5
    else:
      i = 6
    if box[i]==0:
      ball=box
      break
  for box in t:
    if box[i]==2:
      goal=box
      break
  for box in t:
    if box[i] == 5:
      A.append(box)
    elif box[i] == 6:
      B.append(box)

  num_players = len(A) + len(B)

  if len(ball) and len(goal):
    ball_xc = ball[0] + 0.5* (ball[2] - ball[0])
    ball_yc = ball[1] + 0.5* (ball[3] - ball[1])
    goal_xc = goal[0] + 0.5* (goal[2] - goal[0])
    goal_yc = goal[1] + 0.5* (goal[3] - goal[1])

    dis = ( ( goal_xc - ball_xc )**2 + ( goal_yc - ball_yc )**2 )**0.5
    if dis >= 0 and dis <= 100:
      goals_count += 1
      count = 1
      for i in A:
        i_xc = i[0] + 0.5* (i[2] - i[0])
        i_yc = i[1] + 0.5* (i[3] - i[1])
        dis = ( ( goal_xc - i_xc )**2 + ( goal_yc - i_yc )**2 )**0.5
        distances.append(dis)
        team.append('A')
      for i in B:
        i_xc = i[0] + 0.5* (i[2] - i[0])
        i_yc = i[1] + 0.5* (i[3] - i[1])
        dis = ( ( goal_xc - i_xc )**2 + ( goal_yc - i_yc )**2 )**0.5
        distances.append(dis)
        team.append('B')

      #print(distances, team)
      distances, team = zip(*sorted(zip(distances, team)))
      #print(distances, team)

      team_A_count = 0
      team_B_count = 0

      closer = int((3/4)*num_players)
      if closer % 2 == 0:
        closer += 1
      team = team[:closer]
      for i in team:
        if i == 'A':
          team_A_count += 1
        else:
          team_B_count += 1
      message = ''
      if team_A_count > team_B_count:
        message = "Goal for team B"
        goal4B += 1
      else:
        message = "Goal for team A"
        goal4A += 1

      #print(closer)
      goals_count
  return(message, goals_count, goal4A, goal4B)
```

```python
countt = 0
goal_frame = []
for i in results:
  message, goals_count, goal4A, goal4B = Goal_Detection(i)
  if message != '':
    goal_frame = i
    print(message, f"Match results: Team_A: {goal4A} __ Team_B: {goal4B}")
```

## Possessioning & Pass Accuracy

```python
possession_TEAM_A = 0
possession_TEAM_B = 0
possession_IN_CONTEXT = 0
sum = 1
que = []
TEAM_A_TOTAL_PASS = 0
TEAM_B_TOTAL_PASS = 0
TEAM_A_CORRECT_PASS = 0
TEAM_B_CORRECT_PASS = 0
TEAM_A_PASS_ACC = 0
TEAM_B_PASS_ACC = 0
last_possession = []

def Possessioning(frame):
  global possession_TEAM_A
  global possession_TEAM_B
  global possession_IN_CONTEXT
  global sum
  global que
  global TEAM_A_TOTAL_PASS
  global TEAM_B_TOTAL_PASS
  global TEAM_A_CORRECT_PASS
  global TEAM_B_CORRECT_PASS
  global TEAM_A_PASS_ACC
  global TEAM_B_PASS_ACC
  global last_possession

  mat = frame.data
  #mat_np = np.array(mat)
  ball = []
  near_obj = []
  min_dis = 1000000

  if len(mat) == 0:
    return (possession_TEAM_A / sum *100), (possession_TEAM_B / sum *100), (possession_IN_CONTEXT / sum *100), TEAM_A_TOTAL_PASS, TEAM_A_PASS_ACC, TEAM_B_TOTAL_PASS, TEAM_B_PASS_ACC

  i = 6
  for obj in mat:
    if len(obj) == 6:
      i = 5
    if obj[i] == 0:
      ball = obj
      break

  if len(ball) == 0:
    return (possession_TEAM_A / sum *100), (possession_TEAM_B / sum *100), (possession_IN_CONTEXT / sum *100), TEAM_A_TOTAL_PASS, TEAM_A_PASS_ACC, TEAM_B_TOTAL_PASS, TEAM_B_PASS_ACC
  ball_xc = ball[0] + 0.5* (ball[2] - ball[0])
  ball_yc = ball[1] + 0.5* (ball[3] - ball[1])
  for obj in mat:
    obj_xc = obj[0] + 0.5* (obj[2] - obj[0])
    obj_yc = obj[1] + 0.5* (obj[3] - obj[1])

    dis = ( ( obj_xc - ball_xc )**2 + ( obj_yc - ball_yc )**2 )**0.5
    if dis == 0:
      continue
    elif dis < min_dis:
      min_dis = dis
      near_obj = obj


  if min_dis < 100:
    if near_obj[i] == 5:
      possession_TEAM_A += 1

      if len(last_possession) < 5:
        last_possession.append('TA')
      elif len(last_possession) ==5:
        last_possession.pop(0)
        last_possession.append('TA')
      if len(que) ==0:
        que.append('TA')
      elif len(que) ==1:

        if que[0] == 'TB':
          TAcount=last_possession.count('TA')
          TBcount=last_possession.count('TB')
          if TAcount > TBcount:
            que.pop()
            que.append('TA')

      elif len(que)==2:
        if que[0]=='TA':
          TEAM_A_TOTAL_PASS += 1
          TEAM_A_CORRECT_PASS += 1
        elif que[0]=='TB':
          TEAM_B_TOTAL_PASS += 1
        que.pop()
        que.pop()
        que.append('TA')


    elif near_obj[i] == 6:
      possession_TEAM_B += 1
      if len(last_possession) < 5:
        last_possession.append('TB')
      elif len(last_possession) ==5:
        last_possession.pop(0)
        last_possession.append('TB')

      if len(que) ==0:
        que.append('TB')
      elif len(que) ==1:

        if que[0] == 'TA':
          TAcount=last_possession.count('TA')
          TBcount=last_possession.count('TB')
          if TBcount > TAcount:
            que.pop()
            que.append('TB')

      elif len(que) == 2:
        if que[0] == 'TB':
          TEAM_B_TOTAL_PASS += 1
          TEAM_B_CORRECT_PASS += 1
        elif que[0] == 'TA':
          TEAM_A_TOTAL_PASS += 1
        que.pop()
        que.pop()
        que.append('TB')


    else:
      possession_IN_CONTEXT +=1
      if len(que) ==1:
        que.append('IC')
  else:
    possession_IN_CONTEXT += 1
    if len(que) ==1:
      que.append('IC')


  #Calculate the pass accuracy
  if TEAM_A_TOTAL_PASS == 0:
    TEAM_A_PASS_ACC=0
  else:
    TEAM_A_PASS_ACC =TEAM_A_CORRECT_PASS/TEAM_A_TOTAL_PASS*100

  if TEAM_B_TOTAL_PASS ==0:
    TEAM_B_PASS_ACC=0
  else:
    TEAM_B_PASS_ACC =TEAM_B_CORRECT_PASS/TEAM_B_TOTAL_PASS*100


  sum = possession_TEAM_A + possession_TEAM_B + possession_IN_CONTEXT
  #print('The possession of Team A is : ',possession_TEAM_A/sum *100,'% and Team B is : ',possession_TEAM_B / sum *100,'% and In Context : ',possession_IN_CONTEXT / sum *100,'%')
  return (possession_TEAM_A / sum *100), (possession_TEAM_B / sum *100), (possession_IN_CONTEXT / sum *100), TEAM_A_TOTAL_PASS, TEAM_A_PASS_ACC, TEAM_B_TOTAL_PASS, TEAM_B_PASS_ACC

```

```python
for frame in results:
  PA, PB, PC, ATP, APC, BTP, BPC = Possessioning(frame)

print(f'The possession of Team A is : {PA:.2f}, The possession of Team B is : {PB:.2f}, The possession of Team C is : {PC:.2f}')
print(f'The total passes of TEAM A is : {ATP}, and the pass accuracy of TEAM A is : {APC}')
print(f'The total passes of TEAM B is : {BTP}, and the pass accuracy of TEAM B is : {BPC}')
```

```python
import cv2

class HeatMap:
  map = cv2.imread('/content/modified.jpg')

  def __init__(self, frame_count_threshold=30):
      HeatMap.map = cv2.imread('/content/modified.jpg')
      # video.avi
      play_list_name=['SHCC','FHCC','18YL','18YLC','FHF','5YL','SHF','18YRC','18YR','5YR']
      play_list_point=[[290,190],[250,170],[90,180],[135,190],[165,180],[70,160],[375,200],[405,180],[450,160],[470,190]]
      p1=[]
      p2=[]
      self.players_info = None
      for c in range(4):
        ind = play_list_name.index(f_obj_list[c])
        p1.append([f_obj_xc_l[c], f_obj_yc_l[c]])
        p2.append(play_list_point[ind])


      self.pts1 = np.float32(p1)
      self.pts2 = np.float32(p2)
      self.count = -1
      self.frame_count_threshold = frame_count_threshold

  def get_xyxy(self, frame, ID):
    if frame.id == None:
      return None
    for index, id in enumerate(frame.id):
      if id == ID:
        return frame.xyxy[index]

  def heatMap(self, frame, player_id):
      self.count += 1
      if self.count % self.frame_count_threshold != 0:
          return self.map
      self.players_info = self.get_xyxy(frame, player_id)
      #print("Player Info", self.players_info)
      if self.players_info == None:
        return self.map
      (x, y, w, h) = self.players_info

      #print(self.pts1.T, "\n",self.pts2.T)
      matrix = cv2.getPerspectiveTransform(self.pts1, self.pts2)

      pts3 = np.float32([[x + w, y + h]])
      pts3o = cv2.perspectiveTransform(pts3[None, :, :], matrix)
      x1 = int(pts3o[0][0][0])
      y1 = int(pts3o[0][0][1])
      pp = (x1-7, y1-7)
      pp2 = (x1+7, y1+7)
      image = self.map.copy()
      cv2.rectangle(image, pp, pp2, [0, 0, 200], -1)
      alpha = 0.4
      image_new = cv2.addWeighted(image, alpha, self.map, 1 - alpha, 0)
      cv2.imwrite("/content/modified.jpg", image_new)
      self.map = image_new
      return self.map
```

```python
f_obj_list = []
f_obj_xc_l = 0
f_obj_yc_l = 0


def get_heat_map(frame_stadium, frame_player, ID = 1):
  global f_obj_list
  global f_obj_xc_l
  global f_obj_yc_l
  flag = False
  mat = frame_stadium.data.cpu()
  mat = np.array(mat)
  obj_list=[]
  obj_l=[]
  for obj in mat:
    if (obj[5] == 3) and ('FHCC' not in obj_list):
      obj_list.append('FHCC')
      obj_l.append(obj)
    elif (obj[5] == 4) and ('FHF' not in obj_list):
      obj_list.append('FHF')
      obj_l.append(obj)
    elif (obj[5]==5) and ('SHCC' not in obj_list):
      obj_list.append('SHCC')
      obj_l.append(obj)
    elif (obj[5]==6) and ('SHF' not in obj_list):
      obj_list.append('SHF')
      obj_l.append(obj)
    elif obj[5] ==0 :
      obj_xc = obj[0] + 0.5* (obj[2] - obj[0])
      obj_yc = obj[1] + 0.5* (obj[3] - obj[1])
      if obj_yc >= 1920/2:
        if '18YRC' not in obj_list:
          obj_list.append('18YRC')
          obj_l.append(obj)
      else:
        if '18YLC' not in obj_list:
          obj_list.append('18YLC')
          obj_l.append(obj)
    elif obj[5]==1:
      obj_xc = obj[0] + 0.5* (obj[2] - obj[0])
      obj_yc = obj[1] + 0.5* (obj[3] - obj[1])
      if obj_yc >= 1920/2:
        if '18YR' not in obj_list:
          obj_list.append('18YR')
          obj_l.append(obj)
      else:
        if '18YL' not in obj_list:
          obj_list.append('18YL')
          obj_l.append(obj)

    elif obj[5]==2:
      obj_xc = obj[0] + 0.5* (obj[2] - obj[0])
      obj_yc = obj[1] + 0.5* (obj[3] - obj[1])
      if obj_yc >= 1920/2:
        if '5YR' not in obj_list:
          obj_list.append('5YR')
          obj_l.append(obj)
      else:
        if '5YL' not in obj_list:
          obj_list.append('5YL')
          obj_l.append(obj)



  # print(obj_list)
  # print(obj_l)
  obj_xc_l=[]
  obj_yc_l=[]
  for o in obj_l:
    obj_xc = o[0] + 0.5* (o[2] - o[0])
    obj_yc = o[1] + 0.5* (o[3] - o[1])
    obj_xc_l.append(obj_xc)
    obj_yc_l.append(obj_yc)
  # print(obj_l)
  # print(obj_xc_l)
  # print(obj_yc_l)
  # print(int(obj_xc_l[0]))
  # break
  f_obj_l=[]
  f_obj_list=[]
  f_obj_xc_l=[]
  f_obj_yc_l=[]

  for i in range(len(obj_l)):

    if obj_l[i][5]==4 or obj_l[i][5]==6:
      if (obj_l[i][2]-obj_l[i][0]) * (obj_l[i][3]-obj_l[i][1]) < 1920*1080*0.35:
        continue
      else:
        f_obj_l.append(obj_l[i])
        f_obj_list.append(obj_list[i])
        f_obj_xc_l.append(obj_xc_l[i])
        f_obj_yc_l.append(obj_yc_l[i])
    else:
      f_obj_l.append(obj_l[i])
      f_obj_list.append(obj_list[i])
      f_obj_xc_l.append(obj_xc_l[i])
      f_obj_yc_l.append(obj_yc_l[i])


  map_temp = cv2.imread('/content/modified.jpg')
  if len(f_obj_l)<4:
    return flag, map_temp
  else:
    Heat_map = HeatMap(frame_count_threshold=30)
    map = Heat_map.heatMap(frame_player, ID)
    flag = True

  return flag, map
```
